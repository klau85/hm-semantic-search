```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
articles = pd.read_csv('data/articles.csv', dtype={'article_id': str})
```

```python
articles = articles.tail(50000)
```

```python
# instantiate plot
ax = plt.axes()

sns.heatmap(articles.isna().transpose(), ax=ax)

plt.ylabel('Columns')
plt.xlabel('missing values')

plt.show()
```

```python
articles['full_description'] = (articles['article_id'].astype(str)
                                + ' ' + articles['prod_name'] + ' of type '
                                + articles['product_type_name'] + ' from group '
                                + articles['product_group_name'] + ', section '
                                + articles['section_name'] + ' and garment group '
                                + articles['garment_group_name'] + '. '
                                + 'Product color is ' + articles['colour_group_name'] + ' and '
                                + 'perceived color is ' + articles['perceived_colour_value_name'] + '. '
                                + articles['detail_desc'].astype(str)
                                )
```

```python
articles.to_csv('data/articles_full_desc.csv', index=False)
```
